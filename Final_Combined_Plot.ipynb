```python
import os
import matplotlib.pyplot as plt
import numpy as np


#path = 'C:/Users/brind/OneDrive/Desktop/MS/Project/Code/GetOldTweets-python-master'

path = 'C:/Users/brind/OneDrive/Desktop/UPDATED_PROJECT'


placeholder1 = "positive_values.txt"
placeholder2 = "negative_values.txt"
placeholder3 = "neutral_values.txt"

filename1 = os.path.join(path, placeholder1)
filename2 = os.path.join(path, placeholder2)
filename3 = os.path.join(path, placeholder3)

fig = plt.figure(figsize = (20, 10))
ax = fig.add_subplot()

def remove():
    os.remove("positive_values.txt")
    os.remove("negative_values.txt")
    os.remove("neutral_values.txt")

def get_folder():
    folder = []
    fol_name = '2020-'
    for x in os.listdir(path):
        if os.path.isdir(x): 
            result = x.startswith(fol_name)
            if result is True:
                folder.append(x)
    #print("folder:", folder)
    length_fol = len(folder)
    #print("Length:", length_fol)
    #return folder
    get_file(folder)


def get_file(folder):
    for fol in folder:
        abc = os.path.join(path, fol)
        #print("abc...", abc)
        for root, dirs, files in os.walk(abc): 
            for file in files:
                if file.startswith('output'): 
                    
                    
                    filename = os.path.join(abc, file)
                    
                    with open(filename, 'r',encoding="utf-8") as filehandle:
                        #print("Printing from file")
                        for line in filehandle:
                           
                            if (line.startswith("Value")):
                                #print("Line:", line)
                                split_line = line.split("(",1)[1]
                                #print("split line", split_line)
                                remove1 = split_line.replace(')', '')
                                remove2 = remove1.replace(',', '')
                                
                                # using split() 
                                # to extract words from string 
                                res = remove2.split()
                                #print("res", res)

                                
                                f1 = open(filename1, "a", encoding="utf-8")
                                f2 = open(filename2, "a", encoding="utf-8")
                                f3 = open(filename3, "a", encoding="utf-8")
                               
                                print(res[0], file = f1)
                                
                                print(res[1], file = f2)
                                
                                print(res[2], file = f3)
                                
                                
                                f1.close()
                                f2.close()
                                f3.close()
                                
                                
    with open(filename1) as p:
        pos = p.read().splitlines()
    pos = list(map(float, pos))
    #print(pos)


    with open(filename2) as n:
        neg = n.read().splitlines()
    neg = list(map(float, neg))
    #print(neg)

    with open(filename3) as nu:
        neut = nu.read().splitlines()
    neut = list(map(float, neut))
    #print(neut)

    x = np.arange(len(folder))  # the label locations
    width = 0.3 # the width of the bars
    
    #fig = plt.figure(figsize = (20, 10))
    #ax = fig.add_subplot()
    rects1 = ax.bar(x + width/9, pos, width, label='Positive', color = 'green')
    rects2 = ax.bar(x + width, neg, width, label='Negative', color = 'red')
    rects3 = ax.bar(x + width*(1.8), neut, width, label='Neutral', color = 'blue')
   

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('Sentiment Percentage', fontsize = 20)
    ax.set_xlabel('Days', fontsize = 20)
    ax.set_title('Days VS Sentiment Percentage for month ', fontsize = 25)
    ax.set_xticks(x)
    
    plot(folder)
    

def plot(folder):
    f = folder
    f = ['01-Feb','02-Feb','08-Feb','09-Feb','15-Feb','16-Feb','22-Feb','23-Feb','07-Mar','08-Mar','14-Mar','15-Mar','21-Mar','23-Mar','28-Mar','30-Mar','04-Apr','05-Apr','11-Apr','12-Apr','18-Apr','19-Apr','25-Apr','26-Apr','02-May','03-May','09-May','10-May','16-May','17-May','23-May','24-May','06-Jun','07-Jun','13-Jun','14-Jun','19-Jun','20-Jun','26-Jun','27-Jun','04-Jul','05-Jul','10-Jul','11-Jul','12-Jul','17-Jul','18-Jul','19-Jul']

    print(f)
    
    ax.set_xticklabels(f, rotation='vertical')
    
    ax.legend()
    
    plt.legend(loc = 'upper right')

    #autolabel(rects1)
    #autolabel(rects2)
    #autolabel(rects3)

    #fig.tight_layout()
    plt.tight_layout()

    plt.show()
    
   

                                
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 4),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom') 
                                

############################## main ######################################



get_folder()
remove()


```
