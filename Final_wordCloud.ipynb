```python
import os
import matplotlib.pyplot as plt
import numpy as np
from wordcloud import WordCloud


#path = 'C:/Users/brind/OneDrive/Desktop/MS/Project/Code/GetOldTweets-python-master'

path = 'C:/Users/brind/OneDrive/Desktop/UPDATED_PROJECT'
    
lst = []


def get_folder(name):
    folder = []
    fol_name = '2020-' + name
    for x in os.listdir(path):
        if os.path.isdir(x): 
            result = x.startswith(fol_name)
            if result is True:
                folder.append(x)
    #print("folder:", folder)
    length_fol = len(folder)
    get_file(folder)

def get_file(folder):
    for fol in folder:
        abc = os.path.join(path, fol)
        for root, dirs, files in os.walk(abc): 
            for file in files:
                if file.startswith('df_lst.txt'): 
                    
                    filename = os.path.join(abc, file)
                    #print("\n")
                    
                    with open (filename, 'r', encoding = 'utf-8') as filehandle:
                        data = filehandle.read()
                        
                        file_tokens = "tokens"+ name + "." "txt"
                        #print("filetokens....", file_tokens)
                        fileio = open(file_tokens,'a', encoding = 'utf-8')
       
                        fileio.write(data+"\n")  
                        #print("DONE WRITING TO FILE.......")
            
            
                        
def wordcloud():

    file_tokens = "tokens"+ name + "." "txt"
    title = "Word Cloud for the month" + " " + name
    #print("filetokens in wordcloud....", file_tokens)
    with open (file_tokens, 'r', encoding = 'utf-8') as filehandle:
        reading = filehandle.read()
        word = reading.replace("'", '')
        word = word.replace("stock market", '')
        wordcloud = WordCloud(width=900, height=500, max_font_size=150, relative_scaling=0.5,
                      colormap='RdYlGn',
                      normalize_plurals=True).generate(str(word))

        plt.figure(figsize=(10, 7))
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis('off')
        plt.title(title, fontsize = 20)  
        plt.show()
        plt.close()
        
        
def wordcloud_all():
    title = "Total Word Cloud"
    #print("filetokens in wordcloud....", file_tokens)
    filename = "tokens.txt"
    with open (filename, 'r', encoding = 'utf-8') as filehandle:
        reading = filehandle.read()
        #word = reading
        word = reading.replace("'", '')
        word = word.replace("stock market", '')
        wordcloud = WordCloud(width=2000, height=800, max_font_size=150, relative_scaling=0.5,
                      colormap='YlGnBu',
                      normalize_plurals=True).generate(str(word))

        plt.figure(figsize=(20, 10))
        plt.imshow(wordcloud, interpolation="bilinear")
        plt.axis('off')
        plt.title(title, fontsize = 20)  
        plt.show()
        plt.close()

############################## main ######################################

name = '02'
get_folder(name)
wordcloud()

name = '03'
get_folder(name)
wordcloud()

name = '04'
get_folder(name)
wordcloud()

name = '05'
get_folder(name)
wordcloud()

name = '06'
get_folder(name)
wordcloud()

name = '07'
get_folder(name)
wordcloud()

tokens = ["tokens02.txt", "tokens03.txt", "tokens04.txt", "tokens05.txt", "tokens06.txt", "tokens07.txt"]
for t in tokens:
    print(t)
    with open (t, 'r', encoding = 'utf-8') as filehandle:
        data = filehandle.read()
                        
        file = "tokens.txt"
        #print("filetokens....", file_tokens)
        fileio = open(file,'a', encoding = 'utf-8')
       
        fileio.write(data+"\n") 
        
wordcloud_all()


```
